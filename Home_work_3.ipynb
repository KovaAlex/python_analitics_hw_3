–ï—Å–ª–∏ —Ä–µ–Ω–æ–≤–∞—Ü–∏–∏ –¥–æ–º–∞ –Ω–µ –±—ã–ª–æ, —Ç–æ –≤ –Ω–æ–≤–æ–º –ø—Ä–∏–∑–Ω–∞–∫–µ –ø–æ—Å—Ç–∞–≤—å—Ç–µ 0


1.3 –°–æ–∑–¥–∞–π—Ç–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –≥–æ–¥–∞ –ø—Ä–æ–¥–∞–∂–∏, –º–µ—Å—è—Ü–∞ –ø—Ä–æ–¥–∞–∂–∏

1.4 –£–¥–∞–ª–∏—Ç–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ date, zipcode, lat, long

```python
df = pd.read_csv('./kc_house_data.csv')
df.head()
```

### 1.1 –°–æ–∑–¥–∞—Ç—å –Ω–æ–≤—ã–π –ø—Ä–∏–∑–Ω–∞–∫ price_per_sq_lot, –∫–æ—Ç–æ—Ä—ã–π –±—É–¥–µ—Ç —Å–æ–¥–µ—Ä–∂–∞—Ç—å —Å—Ç–æ–∏–º–æ—Å—Ç—å –∑–∞ –æ–¥–∏–Ω –∫–≤. –º–µ—Ç—Ä –æ–±—â–µ–π –ø–ª–æ—â–∞–¥–∏

```python
price_sqft = df['price']/df['sqft_lot']
df['price_per_sq_lot'] = price_sqft
df.head()
# –∑–∞–¥–∞–Ω–∏—è 1.1 –Ω–µ—Ç –≤ –ø—Ä–æ–≥—Ä–∞–º–º–µ, –Ω–æ –Ω–∞ —Å–µ–º–∏–Ω–∞—Ä–µ –≤—ã –ø—Ä–æ –Ω–µ–≥–æ –≥–æ–≤–æ—Ä–∏–ª–∏, —Ä–µ—à–∏–ª —Å–¥–µ–ª–∞—Ç—å.
```

### 1.2 –°–æ–∑–¥–∞—Ç—å –Ω–æ–≤—ã–π –ø—Ä–∏–∑–Ω–∞–∫ delta_renovated, –∫–æ—Ç–æ—Ä—ã–π –±—É–¥–µ—Ç —Å–æ–¥–µ—Ä–∂–∞—Ç—å —Ä–∞–∑–Ω–∏—Ü—É –≤ –≥–æ–¥–∞—Ö –º–µ–∂–¥—É –≥–æ–¥–æ–º —Ä–µ–Ω–æ–≤–∞—Ü–∏–µ–π –¥–æ–º–∞ –∏ –≥–æ–¥–æ–º –ø–æ—Å—Ç—Ä–æ–π–∫–∏ –¥–æ–º–∞

* –ï—Å–ª–∏ —Ä–µ–Ω–æ–≤–∞—Ü–∏–∏ –¥–æ–º–∞ –Ω–µ –±—ã–ª–æ, —Ç–æ –≤ –Ω–æ–≤–æ–º –ø—Ä–∏–∑–Ω–∞–∫–µ –ø–æ—Å—Ç–∞–≤—å—Ç–µ 0

```python
# 1 –≤–∞—Ä–∏–∞–Ω—Ç –∫–∞–∫ –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å —á–µ—Ä–µ–∑ lambda
df['delta_renovated'] = (df['yr_renovated'] - df['yr_built']).apply(lambda x: x if x > 0 else 0)
df.sample(10)
```

```python
df[['yr_built', 'yr_renovated', 'delta_renovated']].sample(14)
```

```python
# 2 –≤–∞—Ä–∏–∞–Ω—Ç —Ü–∏–∫–ª —á–µ—Ä–µ–∑ –º–µ—Ç–æ–¥ iterrow
yr_delta = []
for i, row in df.iterrows():
    if row['yr_renovated']:
        yr_delta.append(row['yr_renovated']-row['yr_built'])
    else:
        yr_delta.append(0)
        
df['delta_renovated'] = yr_delta

df[['yr_built', 'yr_renovated', 'delta_renovated']].head(10)
```

### 1.3 –°–æ–∑–¥–∞–π—Ç–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –≥–æ–¥–∞ –ø—Ä–æ–¥–∞–∂–∏, –º–µ—Å—è—Ü–∞ –ø—Ä–æ–¥–∞–∂–∏

```python
df['date'] = pd.to_datetime(df['date'])
df['Month_Sale'] = df['date'].dt.month_name()
df['Year_Sale'] = df['date'].dt.year
df[['date', 'Month_Sale', 'Year_Sale']].sample(10)
```

### 1.4 –£–¥–∞–ª–∏—Ç–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ date, zipcode, lat, long

```python
df.drop(columns=['date', 'zipcode', 'lat', 'long'], inplace=True)
df.head()
```

–û—Ç—Å–æ—Ä—Ç–∏—Ä—É–π—Ç–µ –æ—Ç –º–µ–Ω—å—à–µ–π —Å—Ç–æ–∏–º–æ—Å—Ç–∏ –∫ –±–æ–ª—å—à–µ–π

```python
# 1 –≤–∞—Ä–∏–∞–Ω—Ç
df.groupby('bedrooms')['price'].mean().sort_values()
```

```python
# 2 –≤–∞—Ä–∏–∞–Ω—Ç

df.groupby('bedrooms').agg({'price': 'mean'}).sort_values('price')
```

### 3.2 –ù–∞–π–¥–∏—Ç–µ –º–∏–Ω–∏–º–∞–ª—å–Ω—É—é, —Å—Ä–µ–¥–Ω—é—é –∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—É—é —Å—Ç–æ–∏–º–æ—Å—Ç–∏ –¥–æ–º–æ–≤ –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç —Å–æ—Å—Ç–æ—è–Ω–∏—è –¥–æ–º–∞

```python
df.groupby('condition').agg({'price': ['min', 'mean', 'max']}).round()
```

### 3.3 –ü–æ—Å—Ç—Ä–æ–π—Ç–µ —Ç–∞–±–ª–∏—Ü—É —Å –ø–æ–¥—Å—á–µ—Ç–æ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –¥–æ–º–æ–≤ –≤ –¥–∞–Ω–Ω—ã—Ö –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –≤–∏–¥–∞ –Ω–∞ –Ω–∞–±–µ—Ä–µ–∂–Ω—É—é –∏ –æ—Ü–µ–Ω–∫–æ–π –≤–∏–¥–∞

```python
df.pivot_table(index='view',
              columns='waterfront',
               values='id',
              aggfunc='count',
              fill_value=0,
              margins=True)
```

### 3.4 –ö–∞–∫–∏—Ö –¥–æ–º–æ–≤ –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç —ç—Ç–∞–∂–Ω–æ—Å—Ç–∏ –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —Å–ø–∞–ª–µ–Ω –±–æ–ª—å—à–µ?

```python
# 1 —Å–ø–æ—Å–æ–± —á–µ—Ä–µ–∑ pivot
df.pivot_table(index='floors',
              columns='bedrooms',
               values='id',
              aggfunc='count',
              fill_value=0,
              margins=True)
```

–ë–æ–ª—å—à–µ –æ–¥–Ω–æ—ç—Ç–∞–∂–Ω—ã—Ö –¥–æ–º–æ–≤ —Å 3-–º—è —Å–ø–∞–ª—å–Ω—è–º–∏ –∏ –≤ —Ü–µ–ª–æ–º –¥–æ–º–æ–≤ —Å 3-–º—è —Å–ø–∞–ª—å–Ω—è–º–∏ –±–æ–ª—å—à–µ –≤—Å–µ–≥–æ

```python
# 2 —Å–ø–æ—Å–æ–± —á–µ—Ä–µ–∑ crosstab
pd.crosstab(index=df['floors'],
           columns=df['bedrooms'],
           margins=True)
# –≤—Å–µ —Å–æ–≤–ø–∞–ª–æ, –ø—Ä–æ—Å—Ç–æ crosstab –ø–∏—Å–∞—Ç—å –º–µ–Ω—å—à–µ 
```

### 3.5 –ü–æ—Å—Ç—Ä–æ–π—Ç–µ —Ç–∞–±–ª–∏—Ü—É —Å –ø–æ–¥—Å—á–µ—Ç–æ–º –º–µ–¥–∏–∞–Ω–Ω–æ–π —Å—Ç–æ–∏–º–æ—Å—Ç–∏ –¥–æ–º–æ–≤ –≤ –¥–∞–Ω–Ω—ã—Ö –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç —Å–æ—Å—Ç–æ—è–Ω–∏—è –¥–æ–º–∞ –∏ –æ—Ü–µ–Ω–∫–∏ –¥–æ–º–∞
